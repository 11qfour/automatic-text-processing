*Задание 1.5*

Сделайте лемматизиацию новостных текстов и сохраните полученные тексты, например в новый столбец датафрейма с датасетом(они пригодятся в части 3 этой лабораторной работы).

!!! Заметьте, что в `pandas` можно сохранять датафрейм в файл (файл из `colab` можно скачать из раздела `Files`). Это удобно, чтобы исключить повторную обработку при перезапуске сессии в `colab`.

````python
from pymorphy3 import MorphAnalyzer
import re
from tqdm.auto import tqdm

tqdm.pandas(desc="Обработка текста")

morph = MorphAnalyzer()

def restore_proper_noun_case(lemma, original_word):
    """
    Восстанавливает регистр для имен собственных
    """
    # Если исходное слово было с заглавной буквы
    if original_word.istitle() or original_word.isupper():
        # Для имен собственных возвращаем заглавную первую букву
        if lemma and len(lemma) > 0:
            # Проверяем, является ли слово именем собственным
            parsed = morph.parse(original_word)
            if parsed:
                tags = parsed[0].tag
                # Если это имя собственное (существительное, собственное)
                if 'Name' in tags or 'Surn' in tags or 'Patr' in tags or 'Geox' in tags:
                    return lemma.capitalize()
                # Если исходное слово было полностью в верхнем регистре
                if original_word.isupper():
                    return lemma.upper()
                # Если начиналось с заглавной буквы в середине предложения
                if original_word.istitle():
                    return lemma.capitalize()
    return lemma

def preprocess_text_pymorphy_with_case(text: str, morph_analyzer: MorphAnalyzer) -> str:
    """
    Лемматизация с сохранением регистра имен собственных
    """
    # Очищаем текст от лишних символов, но сохраняем регистр
    text = re.sub(r'[^а-яА-ЯёЁa-zA-Z\s]', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()

    tokens = text.split()
    lemmatized_tokens = []

    for word in tokens:
        # Получаем лемму
        try:
            parsed = morph_analyzer.parse(word)
            if parsed:
                lemma = parsed[0].normal_form
                # Восстанавливаем регистр для имен собственных
                lemma = restore_proper_noun_case(lemma, word)
            else:
                lemma = word
        except:
            lemma = word

        lemmatized_tokens.append(lemma)

    return " ".join(lemmatized_tokens)

df['processed_pymorphy_case'] = df['text'].progress_apply(
    lambda text: preprocess_text_pymorphy_with_case(text, morph)
)


````
